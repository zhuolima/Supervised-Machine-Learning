Which model does a better job on predicting?

Random Tree Model has a higher true and false positive rate as the number of dependent variable increases in a dataset.

Logistic Regression Model: Performs better with independent variables less or equal to the numbers of dependent variable.

Education Guess: In this dataset, there is only one dependent variable, use logistic regression produce a more accurate model.




Conclusion: Scaler inproves the Logistic Regression Model and do a little or same affect on RandomTree Model. 

```python
import numpy as np
import pandas as pd
from pathlib import Path
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectFromModel
```

```python
train_df = pd.read_csv(Path('Resources/Generator/2019loans.csv'))
test_df = pd.read_csv(Path('Resources/Generator/2020Q1loans.csv'))

X = train_df.drop(['target'], axis=1)
X


```

```python
X_dummies = pd.get_dummies(X)
print(X_dummies.columns)
X_dummies
```

```python
# Converting output labels to 0 and 1


y = test_df.drop(['target'], axis=1)
y_dummies = pd.get_dummies(y)
print(y_dummies.columns)
y_dummies
```

```python
# Convert categorical data to numeric and separate target feature for training data

y_label = LabelEncoder().fit_transform(train_df['target'])
y_label

y_label_test = LabelEncoder().fit_transform(test_df['target'])
y_label_test
```

```python
# add missing dummy variables to testing set

for col in X_dummies.columns:
    if col not in y_dummies.columns:
        y_dummies[col]=0
```

```python
print("Shape: ", X_dummies.shape, y_dummies.shape)
```

```python
# Train the Logistic Regression model on the unscaled data and print the model score

from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier

classifier.fit(X_dummies, y_label)

print(f"Training Data Score: {classifier.score(X_dummies, y_label)}")
print(f"Testing Data Score: {classifier.score(y_dummies, y_label_test)}")
```

```python
# Train a Random Forest Classifier model and print the model score

clf = RandomForestClassifier(random_state=1, n_estimators=500).fit(X_dummies, y_label)
print(f'Training Score: {clf.score(X_dummies, y_label)}')
print(f'Testing Score: {clf.score(y_dummies, y_label_test)}')
```

```python
# Scale the data
scaler = StandardScaler().fit(X_dummies)
X_train_scaled = scaler.transform(X_dummies)
X_train_scaled

X_test_scaled = scaler.transform(y_dummies)
X_test_scaled
```

```python
# Train the Logistic Regression model on the scaled data and print the model score
clf = LogisticRegression().fit(X_train_scaled, y_label)
print(f'Training Score: {clf.score(X_train_scaled, y_label)}')
print(f'Testing Score: {clf.score(X_test_scaled, y_label_test)}')
```

```python
# Train a Random Forest Classifier model on the scaled data and print the model score
clf = RandomForestClassifier(random_state=1, n_estimators=500).fit(X_train_scaled, y_label)
print(f'Training Score: {clf.score(X_train_scaled, y_label)}')
print(f'Testing Score: {clf.score(X_test_scaled, y_label_test)}')
```
